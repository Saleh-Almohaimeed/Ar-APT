```python
import json
import pandas as pd
import re
import openai
import time

from openai import OpenAI


# Create an OpenAI client with your deepinfra token and endpoint
openai = OpenAI(
    api_key="Enter Your API KEY",
    base_url="https://api.deepinfra.com/v1/openai",
)

# ################ Promting Funtion #####################
def ragPrompt(messages):

  resp = openai.chat.completions.create(
    model="Model Name",
    messages= messages
)


  result = resp.choices[0].message.content
  return result


def craftPrompt():
        prompt = '''
           Determine if the text is generated by AI or by a human, Replay only by AI or Human,
          - Note: Give the requested answer without any explaination.
          '''
        return prompt
```

```python
import json
# Load your JSON file
with open('/content/Input.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

i = 0
results = []
for item in data:
    if item['J_similarity'] > 0.05:
      i = i + 1
      messages = []
      messages = [
            {"role": "system", "content": "You are a helpful assistant who always determine if Arabic text is generated by AI or by a Human. Only respond with Human or AI"},
        ]

      prompt = craftPrompt()
      prompt = prompt + "\n" + item['Polish_text']
      messages.append({"role": "user", "content": prompt})
      try:
          output = ragPrompt(messages)
      except:
          continue
      print(i)
      print(output)
      print("==============")

        new_entry = {
                  "original": item['original'],
                  "AI/Human": output,
                  "Polish_text": item['Polish_text'],
                  "Polish_Percentage": item['Polish_Percentage'],
                  "domain": item['domain'],
                  "J_similarity": item['J_similarity']
              }
        results.append(new_entry)
      json_file = "/content/Output.json"
      results.append(new_entry)


with open(json_file, "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print("âœ… Result appended to model_responses.json")
```
